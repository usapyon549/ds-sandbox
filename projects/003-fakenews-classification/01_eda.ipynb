```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1.データ読み込み

```python
url = "https://github.com/tanreinama/Japanese-Fakenews-Dataset/raw/refs/heads/master/fakenews.csv"

df_raw = pd.read_csv(url)
```

```python
df_raw.head(5)
```

・カラムの定義
id	ユニークID  
context	記事の文章（UTF-8エンコード）  
isfake	記事がフェイクであるかどうかのタグ：  
    0：オリジナルの記事  
    1：部分的にフェイク  
    2：完全にフェイク  
nchar_real	記事の中の、人間が執筆した部分の文字数  
ncahr_fake	記事の中の、モデルが生成した部分の文字数  

## 2. データの分布の確認

```python
df_eda = df_raw.copy()
```

```python
# 目的変数isfakeの分布の確認
# 0：オリジナルの記事
# 1：部分的にフェイク
# 2：完全にフェイク

data = df_eda["isfake"].value_counts().to_frame().reset_index()
sns.barplot(data=data,
           x="isfake",
            y= "count"
           )
plt.xlabel("isfake")
plt.ylabel("Frequency")
plt.title("Barplot of isfake")
plt.show()
```

```python
# contextの文章の長さの分布

df_eda["context_len"] = df_eda["context"].str.len()
```

```python
# data = df_eda["context_len"].value_counts().to_frame().reset_index()
# sns.barplot(data=data,
#            x="context_len",
#             y= "count"
#            )

sns.histplot(data=df_eda["context_len"].to_frame(), x="context_len")

plt.xlabel("context_len")
plt.ylabel("Frequency")
plt.title("Histogram of context_len")
plt.show()
```

```python
# contextの何割がfakeか
df_eda["fake_ratio"] = df_eda["nchar_fake"]/df_eda["context_len"]
```

```python
sns.histplot(data=df_eda["fake_ratio"].to_frame(), x="fake_ratio")

plt.xlabel("fake_ratio")
plt.ylabel("Frequency")
plt.title("Histogram of fake_ratio")
plt.show()
```

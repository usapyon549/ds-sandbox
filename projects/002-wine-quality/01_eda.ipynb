```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from ucimlrepo import fetch_ucirepo 
```

# 1.データセット読み込み

使用するデータセット
https://archive.ics.uci.edu/dataset/186/wine+quality

```python
# fetch dataset 
wine_quality = fetch_ucirepo(id=186) 
```

```python
# datasetの特徴量＋予測対象を一括で読み込む
df_raw = wine_quality.data.original
```

```python
df_raw
```

# 2.データの中身の確認

```python
df_eda = df_raw.copy()
```

```python
# データ型の確認
df_eda.dtypes
```

```python
col_names = df_eda.columns
```

```python
col_names
```

```python
y_names = ["quality", "color"]
```

公式より「Missing Values」がnoであるため、欠損値の確認は割愛

## 2.1 化学成分の分布を確認

```python
# 化学成分のヒストグラムの表示

for col in col_names:
    if col not in y_names:
        sns.histplot(data=df_eda[col].to_frame(), x=col)
        
        # y軸のラベル名を変更
        plt.ylabel("Freqency") 
        plt.title(f"Histogram of {col}")
        plt.show()
```

## 2.2 ターゲットカラムの分布の確認

```python

for col in y_names:
    data = df_eda[col].value_counts().to_frame().reset_index()
    sns.barplot(data=data,
               x=col,
                y= "count"
               )
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.title(f"Barplot of {col}")
    plt.show()
    
```

color→不均衡データ

## 2.3 相関行列のヒートマップを作成

```python
corr_matrix = df_eda.drop(y_names, axis=1).corr()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, 
            annot=True,     # 数値を表示
            cmap="coolwarm", # カラーマップの指定
            vmin=-1, # カラーバーの最小値
            vmax=1,# ・最大値
            fmt=".2f" # 小数点以下の桁数
           )      

plt.title("Correlation Matrix")
plt.show()
```
